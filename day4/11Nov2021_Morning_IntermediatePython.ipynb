```python
# Is-a relationship

class Base:
    def __init__(self, x=-1):
        print("Base.__init__: self= ", self)
        self.m_i = x
        
    def m_fun1(self) :
        print("Base.m_fun1111: self.m_i= ", self.m_i)

    def __del__(self):
        print("Base.__del__: self = ", self)
    
class Derived(Base):
    def __init__(self, x=-1, y=-2):
        print("Derived.__init__: self= ", self)
        self.m_j = y
        Base.__init__(self, x)
    
    def m_fun2(self) :
        print("Derived.m_fun2222: self.m_j= ", self.m_j)
        
    def __del__(self):
        print("Derived.__del__: self = ", self)
        Base.__del__(self)
    
def testBasicInheritance():    
    
    dObj = Derived(10, 20)
    print("= ="*5)
    dObj.m_fun1()
    dObj.m_fun2()
    print("= ="*5)
    
    # del(bObj)
# app
if '__main__' == __name__:   
    testBasicInheritance()
    
    print("-- "*5)
```

```python
# Lib

import abc

class Shape(abc.ABC) :
    @abc.abstractmethod
    def draw(self):
        pass
    
class Triangle(Shape):
    def draw(self):
        print("Triangle drawn in memory")
   

class Rectangle(Shape):
    def draw(self):
        print("Rectangle drawn in memory")
   

# Middle-Ware

def render(obj): # LSP :- Liskov Substitution Principle
    print(" Acquire resources...")
    obj.draw()
    print(" Release Acquired resources...")
    print("-=-"*5)

    
class Circle(Shape): 
    def draw(self):
        print("Circle drawn in memory")

        
class Trapesium (Shape): 
    def draw(self):
        print("Trapesium  drawn in memory")
    
# app

def appMain():
    tObj = Triangle()
    render(tObj)
    
    rObj = Rectangle()
    render(rObj)
    print("-=-"*5)

    cObj = Circle()
    render(cObj)
    
    
    trObj = Trapesium()
    print("trObj= ", trObj)
    render(trObj)
    
if '__main__' == __name__:   
    appMain()
```
